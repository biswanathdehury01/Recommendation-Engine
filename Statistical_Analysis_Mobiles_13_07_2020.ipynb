## Imputing the  price column with median value for the missing observations:

```python
price = mobiles[['price']]
price

from sklearn.impute import SimpleImputer, MissingIndicator
imp = SimpleImputer(missing_values=0, strategy='median')
imp.fit(price)
SimpleImputer()
price = (imp.transform(price))
price = pd.DataFrame(price) 
price.head(5)
```

```python
# Concatinating the Imputed price column with the main dataframe 

mobile = mobiles.drop(['price'], axis=1)
mobile

mobiles_pr = pd.concat([mobile, price], axis=1, sort=False)
mobiles_pr


mobiles_pr = mobiles_pr.set_axis(['asin', 'brand', 'rating_product', 'totalReviews', 'name','rating_user','date','helpfulVote','price'], axis=1, inplace=False)
mobiles_pr.head(5)
```
